```python
# Ho: La duracion promedio de los viajes desde Loop al aeropuerto internacional O'Hare NO cambia los dias sabado lluviosos.
# H1: La duracion promedio de los viajes desde Loop al aeropuerto internacional O'Hare cambia los sabados llvuiosos.

# α = 0.05 nivel de significancia.

# Ho: Generalmente se utiliza con la negativa de la hipotesis a realizar.

# Se utilizara una prueba estadistica t de student
```

```python
# Filtrando solo los dias de la semana

df_3['day_of_week'] = df_3['start_ts'].dt.isocalendar().day
```

```python
# Filtrando solo los dias sabados

saturdays_df = df_3[df_3['day_of_week'] == 6]
```

```python
# Filtrando solo los dias sabados no lluviosos

non_rainy_saturdays = saturdays_df[saturdays_df['weather_conditions'] != 'Bad'] 
```

```python
# Duracion promedio en sabados lluvisos
avg_rainy = rainy_saturdays['duration_seconds'].mean()
print(f"Duración promedio sábados lluviosos: {avg_rainy:.2f} segundos")
```

Gracias a los resultados de los promedios y a las gráficas mostradas podemos empezar a notar que hay una diferencia significativa entre ambos tipos de día
</div>

```python
# Calculo de las desviaciones estandar

std_rainy = rainy_saturdays['duration_seconds'].std()
std_non_rainy = non_rainy_saturdays['duration_seconds'].std()

print(f"Desviación estándar sábados lluviosos: {std_rainy:.2f}")
print(f"Desviación estándar sábados no lluviosos: {std_non_rainy:.2f}")
```

```python
# Tamaños de muestra

n_rainy = 180
n_non_rainy = 888
```

```python
# Cálculo del estadístico t

numerator = avg_rainy - avg_non_rainy
denominator = np.sqrt((std_rainy**2 / n_rainy) + (std_non_rainy**2 / n_non_rainy))
t_statistic = numerator / denominator

print(f"Estadístico t: {t_statistic:.4f}")
```

```python
# Grados de libertad (fórmula de Welch)
s1_sq_n1 = std_rainy**2 / n_rainy
s2_sq_n2 = std_non_rainy**2 / n_non_rainy

df = (s1_sq_n1 + s2_sq_n2)**2 / (s1_sq_n1**2/(n_rainy-1) + s2_sq_n2**2/(n_non_rainy-1))
print(f"Grados de libertad: {df:.2f}")
```

```python
# Valor crítico para α = 0.05 (dos colas)

from scipy import stats

alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)
print(f"Valor crítico t: ±{t_critical:.4f}")
```

```python
# Cálculo del p-value (prueba de dos colas)
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df))
print(f"P-value: {p_value:.10f}")
```

```python
# Conclusiones

Realizado en calculo de P-value, se rechaza al hipotesis nula, por lo que se puede concluir que si hay cambio en el tiempos de duracion promedio de los taxies  en los trayectos de Loop al aeropuerto internacional de Chicago, los sabados lluviosos. Los dias lluvioso se puede decir que se tardara aproximadamente 7 minutos mas que los dias no lluviosos, esto puede servir a las companias de taxies prevenir demoras en el camino afectando a los clientes llegar en tiempo a su destino. Se utilza un nivel de significancia de 0.05 aceptando un 5% de error de cometer un error tipo 1 que es rechazar a hipoteisis nula cuando es verdadera. 
```
